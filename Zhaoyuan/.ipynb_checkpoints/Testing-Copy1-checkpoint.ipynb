```python
def bounds_filter(data,parameters):
    condition1 = data > parameters[0]
    condition2 = data < parameters[1]
    condition = (np.column_stack([condition1,condition2]).all(axis=1))
    return condition
```

```python
lxt = np.array(raw['epicsAll']['LAS_VIT_TIME'])

lxt_nonan = []
for i in range(len(lxt)):
    if not np.isnan(lxt[i]):
        lxt_nonan.append(lxt[i])
        

lxt_nonan = myround(np.asarray(lxt_nonan),0.00001)

lxt_u = np.unique(lxt_nonan)
#get rid of NaN values
andor_intensities_new = andor_intensities.copy()
for i in range(len(NaN_entries)):
    andor_intensities_new[NaN_entries[i]] = 0
andor_intensities_new = andor_intensities_new[andor_intensities_new!=0]
#set bins, step = difference bewteen first two lxt
if len(lxt_u) == 1:
    d_bins = lxt_u[0]
else:
    dstep = lxt_u[1]-lxt_u[0]
    d_bins = np.arange((lxt_u[0]-(dstep/2)),(lxt_u[-1]+(1*dstep/2)),dstep)
    n_d_bins = len(d_bins)

print(d_bins)
print(dstep)
plt.figure()
for dd in d_bins:
    plt.axhline(dd,c='y')
plt.plot(lxt, '.')
plt.xlabel('shot')
plt.ylabel('delay (lxt)')
plt.suptitle('LXT delay range')

plt.figure()
plt.hist(lxt_nonan, bins = d_bins)
for d in d_bins:
    plt.axvline(d,c='y')
plt.xlabel('shot')
plt.ylabel('delay (lxt)')
plt.suptitle('LXT histogram')

#take shots and assign them to each bin (row=# of bin, column=# of shot)
time_tracker = np.zeros([n_d_bins,len(lxt_nonan)])
for i in range (len(lxt_nonan)):
    for j in range (n_d_bins):
        if j==0 & (lxt_nonan[i] <= d_bins[j] + dstep/2):
            time_tracker[j][i] = andor_intensities_new[i]
        elif j == n_d_bins - 1 & (lxt_nonan[i] > d_bins[j] - dstep/2):
            time_tracker[j][i] = andor_intensities_new[i]
        elif (lxt_nonan[i] > d_bins[j] - dstep/2) & (lxt_nonan[i] <= d_bins[j] + dstep/2):
            time_tracker[j][i] = andor_intensities_new[i]
andor_intensities_time = np.zeros(n_d_bins)
print(n_d_bins)
print(time_tracker.shape)

#get average intensity of each bin
for i in range (n_d_bins):
    print(np.count_nonzero(time_tracker[i]))
for i in range (n_d_bins):
    andor_intensities_time[i] = np.sum(time_tracker[i]) / np.count_nonzero(time_tracker[i])

print(np.sum(time_tracker[0]))
plt.figure()
plt.plot(d_bins, andor_intensities_time)

timestamp = raw['timestamp']
```
